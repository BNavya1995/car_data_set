```python
df=df.drop(columns=['Unnamed: 0'])
df.head(30)
```

```python
print(f"\n Before cleaning {df['brand'].nunique()}")

df['brand']=df['brand'].str.strip()

print(f"\n after cleaning {df['brand'].nunique()}")
print(df['brand'].value_counts())
```

```python
df['seller_type']=df['seller_type'].str.strip().str.title()
df['seller_type'].unique()
```

```python
type_mapo={'Individul':'Individual',
           'Individuall':'Individual',
           'Delaer':'Dealer'
           }

df['seller_type']=df['seller_type'].replace(type_mapo)
df['seller_type'].unique()
```

```python
df.info()
```

```python
df['km_driven']=df['km_driven'].astype(str).str.replace('kms','')
df['km_driven']=df['km_driven'].astype(str).str.replace('km','')
df['km_driven']=df['km_driven'].astype(str).str.replace(',','')
```

```python
df['km_driven'].head(100)
```

```python
df['km_driven']=df['km_driven'].astype(int)
df.info()
```

```python
df.isnull().sum() *100/df.shape[0]
```

```python
before=len(df)
df=df.dropna(subset=['selling_price'])
after=len(df)
print(f"Dropped {before-after} where selling price is  missing ")
```

```python
df.isnull().sum() *100/df.shape[0]
```

```python
engine_missing=df[df['engine'].isna()]['vehicle_age']
engine_present=df[df['engine'].notna()]['vehicle_age']

print(f"Average Vehicle where engine is: {engine_missing.mean():1f} years")
print(f"Average Vehicle where engine is:{engine_present.mean():1f}years")

```

```python
df.isnull().sum() *100/df.shape[0]
```

```python
print("Engine distribution:")
print(f"  Mean:   {df['engine'].mean():.0f} cc")
print(f"  Median: {df['engine'].median():.0f} cc")
```

```python
for cat in ['engine','mileage','max_power']:
  median_value=df[cat].median()
  df[cat]=df[cat].fillna(median_value)
  print(f" filled {cat} with median values")
```

```python
df['mileage']=df['mileage'].astype(str).str.replace('kmpl','').str.replace('kmp','').str.strip()

df['mileage']=pd.to_numeric(df['mileage'],errors='coerce')
df['mileage'].head(100)
```

```python
df.isnull().sum()*100/df.shape[0]
```

```python
seat_mode=df['seats'].mode()[0]
df['seats']=df['seats'].fillna(seat_mode)
df['seats']
```

```python

import pandas as pd
from sklearn.metrics import mean_absolute_error

# Assume df is your cleaned dataset with a 'selling_price' column
# Step 1: Compute the mean selling price
mean_price = df['selling_price'].mean()

# Step 2: Build baseline predictions (same value for all records)
baseline_predictions = [mean_price] * len(df)

# Step 3: Calculate MAE
baseline_mae = mean_absolute_error(df['selling_price'], baseline_predictions)

print("Baseline MAE:", baseline_mae)
```
